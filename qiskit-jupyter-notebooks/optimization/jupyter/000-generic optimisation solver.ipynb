## Generic CPLEX Solver

```python
import docplex.mp
from docplex.mp.model_reader import ModelReader
```

```python
m = ModelReader.read('..\\problemFiles\\802-Production Problem.lp')
```

```python
m.print_information()
```

```python
print(m.export_as_lp_string())
```

```python
results = m.solve()
```

```python
m.report()
```

```python
d1 = m.objective_value
print(d1)
```

```python
m.print_solution()
```

```python
def log_results():
    from docplex.mp.constants import int_probtype_to_string
    filename=''
    cpx = m.get_cplex(do_raise=False)
    if cpx:
        cpx_probtype = cpx.get_problem_type()
        filename=filename+int_probtype_to_string(cpx_probtype)+'-'
    
    entry=m.name+','+int_probtype_to_string(cpx_probtype)+','+str(m.number_of_variables)+','+str(m.objective_value)+','
    for var in results.get_all_values():
        entry = entry+str(var)+'  '
    entry = entry+','+'%.2f'%results.solve_details.time+',,0,0,Classical,CPLEX,'+results.solve_details.status+'\n'
    f=open('..\\stats.csv','a')
    f.write(entry)
    f.close()

log_results()
print(results)
```

```python
import numpy as np

N = 10
b = np.random.rand(10,10)
b_symm = (b + b.T)/2
#b = b/2
np.fill_diagonal(b_symm,1.0)
print (b_symm)

matrix = (N,N)
matrix = np.zeros(matrix)
print(matrix)
value=.01
for i in range(0, N):
    for j in range(0, N):
        if i == j:
            matrix[i][j] = 1
        else:
            matrix[i][j] = value
            value = value+.01

matrix = (matrix + matrix.T)/2
print(matrix)
```
